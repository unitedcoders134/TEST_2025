Q: How do modern operating systems and browsers handle infinite loops or excessively long-running scripts?
1. Operating Systems (OS) ‚Äì Process and Resource Management
Modern operating systems use several mechanisms to manage and mitigate infinite loops or resource-hogging applications:

Preemptive Multitasking:
The OS gives each process a time slice. If a program enters an infinite loop, it won‚Äôt freeze the entire system because the OS can still schedule and run other processes.

CPU Usage Monitoring:
Tools like Task Manager (Windows), Activity Monitor (macOS), or top/htop (Linux) help users see processes consuming abnormal CPU resources. Users can manually terminate them.

Watchdog Timers and Kernel Safeguards:
The OS kernel may implement watchdog timers that reset or kill processes that hang or crash system-critical services.

Example (Windows):
When a program becomes unresponsive due to an infinite loop, Windows may display:
"Program is not responding. You can wait for it to respond or close it."

2. Web Browsers ‚Äì JavaScript Execution Safeguards
Browsers are designed to prevent a single web page from freezing the entire tab or browser. Here's how they handle long-running scripts:

Script Timeout Limits:
Browsers like Chrome and Firefox set execution time limits for JavaScript. If a script runs for too long without yielding control, a warning appears.

Example (Chrome):
"Page Unresponsive ‚Äì You can wait for it to become responsive or exit the page."

Example (Firefox):
"A script on this page may be busy or it may have stopped responding."

Single-Threaded JavaScript:
JavaScript runs on a single thread per tab, so infinite loops block rendering and interaction. Browsers detect these freezes and provide recovery options.

Crash Recovery:
Modern browsers isolate each tab and extension in its own process (multi-process architecture). This means one tab crashing due to an infinite loop won‚Äôt crash the entire browser.

DevTools Detection:
Developers using tools like Chrome DevTools can pause execution or set breakpoints to inspect where a loop or long-running code is stuck.

3. Mobile Operating Systems
iOS/Android:
If an app consumes too much CPU or becomes unresponsive due to a loop, the OS may automatically close it and show a message like:

iOS: "App Name quit unexpectedly."

Android: "App isn't responding. Do you want to close it?"

Conclusion:
Modern systems use timeouts, process isolation, preemptive scheduling, and user alerts to detect and manage infinite loops. These protections ensure one misbehaving script or process doesn‚Äôt compromise the performance or usability of the entire system.

Social Network Analysis (SNA) is the process of investigating social structures through the use of networks and graph theory. It helps understand how individuals (or entities) are connected and how information or influence flows through social systems.

üîó Graph Representation in Social Media
In graph theory:

Nodes (Vertices): Represent individual users or entities.
Edges (Links): Represent relationships between those users (e.g., friendships, follows, messages).
Types of Graphs in SNA:
Directed Graphs: Used when relationships have a direction (e.g., Twitter follows ‚Äî A follows B doesn‚Äôt mean B follows A).
Undirected Graphs: Used when relationships are mutual (e.g., Facebook friends ‚Äî both users have to accept the connection).
Weighted Graphs: Edges can have weights showing the strength of a relationship (e.g., number of messages exchanged).
