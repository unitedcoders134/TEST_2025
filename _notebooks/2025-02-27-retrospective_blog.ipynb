### Introduction

Pawnsy is a full-stack project that implements a real-time ranking system with full CRUD functionality. The project follows a structured frontend-backend integration, where the Flask backend provides API endpoints for data storage, retrieval, and updates, while the JavaScript-based frontend dynamically renders the leaderboard using DOM manipulation.

The project demonstrates key computing principles, including data abstraction, algorithms, networking, and computing impact. Users can interact with the leaderboard by viewing rankings, updating scores, and managing data efficiently.

### Backend Implementation & API Calls

The backend, built using Flask and Flask-RESTful, supports a leaderboard system where player rankings and scores are stored in a SQLAlchemy database model. It exposes API endpoints that allow the frontend to fetch and update leaderboard data dynamically.

### Student-Developed Procedure: Handling Score Updates

This is a procedure that processes user score updates based on game outcomes. The function listens for a POST request at /user_stats/update_score and modifies the database accordingly.
```python
@app.route('/user_stats/update_score', methods=['POST'])
def update_score():
    data = request.get_json()
    name = data.get('name')
    result = data.get('result')

    if not name or result not in ['win', 'loss']:
        return jsonify({'error': 'Valid Name and result (win/loss) are required'}), 400

    user = User.query.filter_by(_name=name).first()
    if not user:
        return jsonify({'error': 'User not found'}), 404

    user_stats = UserStats.query.filter_by(user_id=user.id).first()
    if not user_stats:
        return jsonify({'error': 'User stats not found'}), 404

    if result == 'win':
        user_stats.wins += 1
    else:
        user_stats.losses += 1

    user_stats.update()
    return jsonify({'message': 'User stats updated successfully', 'data': user_stats.read()}), 200

```

✔ Procedural Abstraction: This function abstracts the logic required to process score updates and integrates with the SQLAlchemy model to modify data in a structured way.

### Frontend Implementation & API Integration

The frontend is responsible for dynamically fetching, displaying, and updating leaderboard rankings using JavaScript, the Fetch API, and DOM manipulation.

### ✔ Overall sequence: 
First, pythonURI is defined in the script tag, then a fetch request is made to the backend to retrieve leaderboard data. This data is stored in a JavaScript variable and accessed through the DOM, using a forEach loop to populate the leaderboard dynamically. Buttons on each entry are assigned event listeners to handle CRUD operations dynamically.

### Algorithm for Fetching Leaderboard Data

```js
async function fetchLeaderboard() {
    try {
        const response = await fetch(`${pythonURI}/api/user_stats`);
        const users = await response.json();
        const leaderboardBody = document.getElementById("leaderboard-body");
        leaderboardBody.innerHTML = "";

        users.forEach((user, index) => {
            const row = `<tr>
                <td>${index + 1}</td>
                <td>${user.user_name}</td>
                <td>${user.wins + user.losses}</td>
            </tr>`;
            leaderboardBody.innerHTML += row;
        });
    } catch (error) {
        console.error('Error fetching leaderboard:', error);
    }
}

```
### ✔ Use of a List: The function retrieves a list of all users from the API response, demonstrating data storage and retrieval.
