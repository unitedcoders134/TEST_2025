````python
%% js
// Unit 1 — Variables and Assignments
let tasks = [];  // Array to store tasks
let taskId = 1;  // ID to keep track of task numbers

// Unit 2 — Data Abstraction: Let's use objects to represent a task
function createTask(description, priority) {
    return {
        id: taskId,
        description: description,
        priority: priority,
        completed: false
    };
}

// Unit 3 — Mathematical Expressions: Setting priority levels (high:1, medium:2, low:3)
function setPriority() {
    console.log("Set task priority: 1 = High, 2 = Medium, 3 = Low");
    return parseInt(prompt("Choose priority (1, 2, 3): "));
}

// Unit 4 — Strings and String Operations: Let's allow users to add task descriptions
function addTask() {
    const description = prompt("Enter task description: ");
    const priority = setPriority();
    const task = createTask(description, priority);
    tasks.push(task);
    taskId++;
}

// Unit 5 — Boolean Expressions: Checking whether tasks are completed
function markTaskDone(taskId) {
    for (let task of tasks) {
        if (task.id === taskId) {
            task.completed = true;
            console.log(`Task ${taskId} marked as done!`);
            return;
        }
    }
    console.log("Task not found!");
}

// Unit 6 — Conditionals: Mark task done or delete based on user's choice
function modifyTask(taskId) {
    const choice = prompt("Enter 'done' to mark as completed or 'delete' to remove task: ").toLowerCase();
    if (choice === 'done') {
        markTaskDone(taskId);
    } else if (choice === 'delete') {
        deleteTask(taskId);
    } else {
        console.log("Invalid option!");
    }
}

// Unit 7 — Nested Conditionals: Nested checks for deleting a task
function deleteTask(taskId) {
    for (let i = 0; i < tasks.length; i++) {
        if (tasks[i].id === taskId) {
            if (tasks[i].completed) {
                tasks.splice(i, 1);
                console.log(`Task ${taskId} deleted!`);
            } else {
                console.log("Task not completed yet. Complete it before deleting.");
            }
            return;
        }
    }
    console.log("Task not found!");
}

// Unit 8 — Iterations: Iterate over tasks to show them
function viewTasks() {
    if (tasks.length === 0) {
        console.log("No tasks to show.");
    } else {
        for (let task of tasks) {
            const status = task.completed ? "Done" : "Pending";
            console.log(`ID: ${task.id}, Description: ${task.description}, Priority: ${task.priority}, Status: ${status}`);
        }
    }
}

// Unit 10 — Lists: A list to store tasks is used, and now we add basic functionality.
function main() {
    while (true) {
        console.log("\nOptions: (1) Add Task (2) View Tasks (3) Modify Task (4) Exit");
        const choice = parseInt(prompt("Choose an option: "));
        if (choice === 1) {
            addTask();
        } else if (choice === 2) {
            viewTasks();
        } else if (choice === 3) {
            const taskIdInput = parseInt(prompt("Enter task ID: "));
            modifyTask(taskIdInput);
        } else if (choice === 4) {
            console.log("Exiting Task Manager...");
            break;
        } else {
            console.log("Invalid option, try again!");
        }
    }
}

// Running the Task Manager
main();

````
